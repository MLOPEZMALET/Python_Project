```python
import re
import operator
import pandas as pd
```

```python
text = "Dash apps are composed of two parts. The first part is the of the app and it describes what the application looks like. The second part describes the interactivity of the application and will be covered in the next chapter. Dash provides Python classes for all of the visual components of the application. We maintain a set of components in the dash_core_components and the dash_html_components library but you can also build your own with JavaScript and React.js."
```

```python
new = re.sub(r'[^\w\s]', '', text)
print(new)
liste = new.split()
print(liste)
liste_of_frequence = [liste.count(w) for w in liste]
print(liste_of_frequence)
dictionary = dict(zip(liste, liste_of_frequence))
print(dictionary)
#sorted_dict = sorted(dictionary, key=dictionary.items(), reverse=False)
sorted_x = sorted(dictionary.items(), key=operator.itemgetter(1), reverse=True)
print(sorted_x[0][0])
```
